```python
#map 
#map is used to replace a value

d= {'no': False , 'yes': True}
df['international plan']= df['international plan'].map(d)
```

```python
df.head(2)
```

```python
#dataframe transormatiom


total_calls= df['total day calls'] +df['total eve calls'] +df['total night calls'] + df['total intl calls'] 
#print(total_calls)
df.insert(loc= len(df.columns), column = 'total_calls', value = total_calls)
df.head(2)
```

```python


# Grouping

columns_to_show = ['total day minutes', 'total eve minutes', 'total night minutes']
df.groupby(['churn'])[columns_to_show].describe(percentiles = [])
```

```python
df.groupby(['churn'])[columns_to_show].describe()
```

```python
columns_to_show = ['total day minutes', 'total eve minutes', 'total night minutes']

df.groupby(['churn'])[columns_to_show].agg([np.mean, np.std, np.min, np.max])

```

```python
#Pivot is used to show the feature of dataset
#pivot Tables

df.pivot_table(['total day calls', 'total eve calls', 'total night calls'], ['area code'], aggfunc='median')
```

```python
df.pivot_table(['total day calls', 'total eve calls', 'total night calls'], ['area code'], aggfunc='mean')
```
