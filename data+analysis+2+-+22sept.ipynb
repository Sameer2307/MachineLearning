```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#we use seaborn
import seaborn as sns
sns.set()

#Graphics in SVG
%config InlineBacjend.figure_format='svg'
```

```python
df = pd.read_csv("D:\\hackveda\\telecom_churn.csv")
```

```python
#univariate analysis

#histogram
features = ['total intl calls' ,'total day minutes']
df[features].hist(figsize = (12,4))
```

```python
#Density Plot  -  Smooth transition of histogram
df[features].plot(kind='density', subplots=True, layout=(1,2),figsize=(12,4))
```

```python
#seaborn use  - distplot
sns.distplot(df['total intl calls'])
```

```python
#Boxplot
ax = sns.boxplot(x=df['total intl calls'])
```

```python
#violin plot
sns.violinplot(data = df['total intl calls'])
```

```python
#categorical and binary features
df['churn'].value_counts()
```

```python
#count plot
sns.countplot(x='churn',data = df)
```

```python
#multivariate analysis
# drop non numerical values
numerical = list(set(df.columns)-set(['state','international plan',
                                      'voice mail plan', 'area code', 'churn','customer service calls']))
corr_matrix=df[numerical].corr()
sns.heatmap(corr_matrix)
```

```python
print(df[numerical].corr())
```

```python
#Scatterplot
plt.scatter(df['total day minutes'], df['total night minutes'])
```

```python
#joint plot in seaborn
sns.jointplot(x='total day minutes', y= 'total night minutes', data=df, kind='scatter')
```
